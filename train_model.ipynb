```python
# uploading libraries
import tensorflow as tf  
from tensorflow import keras
from keras import layers, models
import numpy as np
import pandas as pd      
```

```python
# train data
train_data=np.load("data/emnist-byclass-train.npz")
X_train,y_train=train_data['training_images'],np.array(train_data['training_labels']).reshape(-1)
# test data
test_data=np.load("data/emnist-byclass-test.npz")
X_test=test_data['testing_images']
test_y=pd.read_csv("data/pred_results_example.csv")
y_test=np.array(test_y['Category'])
```

```python
# initializing model
model=models.Sequential([
    layers.Input((28,28,1)),
    layers.Conv2D(32,kernel_size=(3,3),activation='relu'),
    layers.MaxPool2D(2,2),
    layers.Conv2D(64,kernel_size=(3,3),activation='relu'),
    layers.MaxPool2D(2,2),
    layers.Flatten(),
    layers.Dense(units=128,activation='relu'),
    layers.Dense(units=62,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='SparseCategoricalCrossentropy',metrics=['accuracy'])
```

```python
# training
model.fit(X_train,y_train,epochs=10,batch_size=128,verbose=2)
```

```python
# evaluation and saving the model for further use
_,test_acc=model.evaluate(X_test,y_test)
print(f"Accuracy :{test_acc*100:.2f}%")
model.save("handwritten_nb.keras")
```
